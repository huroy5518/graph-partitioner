# This is a demo for use GraphPartitioner Library

### Start with creating a graph

```python
from GraphPartitioner import Graph
n = 100
g = Graph(n)
```

### we can adding some edges

```python
import random
enum = random.randint(n, n * n)
edges = dict()
for k in range(enum):
    u = random.randint(1, n)
    v = random.randint(1, n)
    if edges.get((u, v), False):
        continue
    g.add_edge(u, v)
    edges[(u, v)] = True
    edges[(v, u)] = True
```

### Choose a partitioner

```python
from GraphPartitioner import KernighanLinPartitioner, KernighanLinPartitionerFast, LDGPartitioner
```

```python
# Take KernighanLinPartitionerFast for example
# pass a graph to it then part
part = KernighanLinPartitionerFast(g)
part.part(2)
```

### Or you can use numpy meshgrid to generate a meshgrid

```python
import numpy as np
nx, ny = (25, 25)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)

xv, yv = np.meshgrid(x, y)
X = Graph(xv, yv)
part = LDGPartitioner(X)
part.part(2)
```

### for that use numpy meshgrid, you can use part_numpy_mesh
### this will return you a meshgrid partitions

```python
part.part_numpy_mesh(2)
```
